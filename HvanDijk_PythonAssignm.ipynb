```python
## Descriptive Analytics:

### Mean:
print ('Price Mean:')
print (Cars_df['Price'].mean(),'\n')
### Median:
print ('Price Median:')
print (Cars_df['Price'].median(),'\n')
### Mode:
print ('Price Mode:')
print (Cars_df['Price'].mode(),'\n')
### Kurtosis:
print ('Price kurtosis:')
print (Cars_df['Price'].kurtosis(),'\n')
### Quartiles:
print ('Price Quartiles:')
print ('1st Quartile: ', Cars_df['Price'].quantile(0.25))
print ('2nd Quartile: ', Cars_df['Price'].quantile(0.50))
print ('3rd Quartile: ', Cars_df['Price'].quantile(0.75),'\n')
### Range:
print ('Price Range:')
print ('min value: ', Cars_df['Price'].min())
print ('min value: ', Cars_df['Price'].max())
print ('Range: ', Cars_df['Price'].max() - Cars_df['Price'].min())
print ('InterQuartile Range - Q3 less Q1: ', Cars_df['Price'].quantile(0.75) - Cars_df['Price'].quantile(0.25),'\n')
### Standard deviation:
print ('Price Standard Deviation:')
print (Cars_df['Price'].std(),'\n')
### Variance:
print ('Price Variance:')
print (Cars_df['Price'].var(),'\n')
### Coefficiant of Variance:
print ('Price Coefficiant of:')
print (Cars_df['Price'].std()/Cars_df['Price'].mean(),'\n')
### Weighted meam:
print ('Price Weighted meam - Mileage used as weighted measure:')
print (sum(Cars_df['Mileage'] * Cars_df['Price'])/Cars_df['Mileage'].sum(),'\n')
### Geometric meam:
print ('Price Geometric meam:')
print (stats.gmean(Cars_df['Price']),'\n')
### Skewness:
print ('Price skewness:')
print (stats.skew(Cars_df['Price']),'\n')
```

```python
## Graphs

### Heatmap:
print("Price / Mileage Heatmap:",'\n')
dff=Cars_df[['Mileage','Price']]
sns.heatmap(dff.corr(), annot=True)
sns.set(rc={'figure.figsize':(40,40)})
```

```python
## Graphs

### Scatter Plot:
print("Price / Mileage scatter Plot:",'\n')
fig = plt.figure(figsize =(40, 40))
plt.scatter(Cars_df['Price'],Cars_df['Mileage'])
plt.show
```

```python
## Graphs

### Boxplot:
print("Price Box Plot:",'\n')
fig = plt.figure(figsize =(40, 40))
sns.boxplot(y=Cars_df['Price'])
plt.show
```
